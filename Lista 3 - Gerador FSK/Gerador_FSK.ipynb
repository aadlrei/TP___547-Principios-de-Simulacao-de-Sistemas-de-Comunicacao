```python
import numpy as np
import matplotlib.pyplot as plt

# === Configurações ===
fs = 100000  # Frequência de amostragem em Hz (100 kHz)
duration = 0.01  # Duração total do sinal em segundos (10 ms)
t = np.linspace(0, duration, int(fs*duration), endpoint=False)  # Tempo

f1 = 5000  # Frequência para bit 1 (5 kHz)
f2 = 2000  # Frequência para bit 0 (2 kHz)

# === Gerar dados binários aleatórios ===
data = (np.random.rand(10) > 0.5)
samples_per_bit = int(len(t) / len(data))  # Quantas amostras para cada bit
data_sequence = np.repeat(data, samples_per_bit)

# Ajustar tamanho (caso haja pequena diferença)
data_sequence = data_sequence[:len(t)]

# === Gerar portadoras ===
carrier1 = np.sin(2 * np.pi * f1 * t)  # Para bit 1
carrier2 = np.sin(2 * np.pi * f2 * t)  # Para bit 0

# === Gerar sinal FSK ===
fsk = carrier1 * data_sequence + carrier2 * (1 - data_sequence)

# === Plotar ===
plt.figure(1)
plt.plot(t, carrier1)
plt.title('Portadora para bit 1 (f1 = 5 kHz)')
plt.figure(2)
plt.plot(t, carrier2)
plt.title('Portadora para bit 0 (f2 = 2 kHz)')
plt.figure(3)
plt.plot(t, data_sequence)
plt.title('Mensagem Binária Expandida')
plt.figure(4)
plt.plot(t, fsk)
plt.title('Sinal FSK')
plt.show()
```
